# animations — Image → 3D on your own GPU (Colab)

Run TripoSR here on your Colab GPU and expose it as a **public Gradio link**.
Paste that `https://….gradio.live` link into the Studio's **Connect a generator**
field, then upload a photo and Generate — it runs on *your* GPU: fast, no queue,
no per-model cost.

**Steps:** Runtime → Change runtime type → **GPU**, then Runtime → **Run all**.
When the last cell prints a `gradio.live` URL, copy it into the Studio.

> The link is alive only while this notebook runs, and the URL changes each
> session — paste the fresh one into the Studio each time.

```python
# 1) Get TripoSR + dependencies
!git clone --depth 1 https://github.com/VAST-AI-Research/TripoSR.git
%cd TripoSR
!pip install -q -r requirements.txt
!pip install -q gradio  # ensure a recent Gradio for share links
```

```python
# 2) Launch TripoSR's Gradio app with a public share link.
#    We patch launch() to enable share=True, then run it.
import re, pathlib
src = pathlib.Path('gradio_app.py').read_text()
if 'share=True' not in src:
    src = re.sub(r'\.launch\(', '.launch(share=True, ', src, count=1)
    pathlib.Path('gradio_app.py').write_text(src)
print('Starting… watch for a line like:  Running on public URL: https://XXXX.gradio.live')
!python gradio_app.py
```

## Use it
1. Copy the `https://XXXX.gradio.live` URL printed above.
2. Open the Studio: **zurlazar.github.io/animations/studio.html**
3. Paste the URL into **Connect a generator** → **Save** (status turns green).
4. Upload a device photo → **Generate 3D model**.

Keep this notebook running while you use the Studio. Quality tip: a clean,
single, well-lit product shot on a plain background works best.

Want higher quality? Swap TripoSR for **TRELLIS** or **Hunyuan3D** (heavier
install) — the Studio speaks the same Gradio protocol, though those apps expose
a multi-step API and may need a small tweak.
