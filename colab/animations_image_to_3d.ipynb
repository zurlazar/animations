# animations — Image → 3D on your own GPU (Colab)

Run TripoSR here on your Colab GPU and expose it as a **public Gradio link**.
Paste that `https://….gradio.live` link into the Studio's **Connect a generator**
field, then upload a photo and Generate — it runs on *your* GPU: fast, no queue,
no per-model cost.

**Steps:** Runtime → Change runtime type → **GPU**, then Runtime → **Run all**.
When the last cell prints a `gradio.live` URL, copy it into the Studio.

> The link is alive only while this notebook runs, and the URL changes each
> session — paste the fresh one into the Studio each time.

```python
# 1) Get TripoSR + its dependencies (fresh, from /content)
%cd /content
!rm -rf TripoSR
!git clone --depth 1 https://github.com/VAST-AI-Research/TripoSR.git
%cd /content/TripoSR
!pip install -q -r requirements.txt

# Reconcile versions with Colab's ecosystem:
#  - gradio 4.44 (Studio-compatible) + matching gradio-client
#  - huggingface_hub 0.25.2 still has HfFolder (gradio 4.44 imports it)
#  - transformers/tokenizers matched to that hub
#  - onnxruntime for rembg (background removal)
#  - trimesh>=4.5: TripoSR's pin predates numpy 2 (ndarray.ptp() was removed,
#    crashing the final mesh-orientation step after a successful generation)
#  - pydantic 2.10.6: Colab's newer pydantic breaks gradio 4.44's /info route
#    (TypeError: argument of type 'bool' is not iterable)
#  - numpy==2.0.2: the one version Colab's whole chain accepts — cupy needs
#    >=2.0, but numba (pymatting/rembg) needs <=2.0 (TripoSR pins 1.x, too old)
!pip install -q -U "gradio==4.44.1" "gradio-client==1.3.0" "huggingface_hub==0.25.2" "transformers==4.44.2" "tokenizers==0.19.1" onnxruntime "numpy==2.0.2" "pydantic==2.10.6" "trimesh>=4.5"
```

```python
# 2) Higher-quality background removal, then launch with a public share link.
%cd /content/TripoSR
# TripoSR defaults to rembg's u2net; isnet-general-use gives noticeably cleaner
# cutouts, and TripoSR's geometry is very sensitive to the mask quality.
!sed -i 's/rembg.new_session()/rembg.new_session("isnet-general-use")/' gradio_app.py
print('Starting… watch for:  Running on public URL: https://XXXX.gradio.live')
!python gradio_app.py --share
```

## Use it
1. Copy the `https://XXXX.gradio.live` URL printed above.
2. Open the Studio: **zurlazar.github.io/animations/studio.html**
3. Paste the URL into **Connect a generator** → **Save** (status turns green).
4. Upload a device photo → **Generate 3D model**.

Keep this notebook running while you use the Studio. Quality tip: a clean,
single, well-lit product shot on a plain background works best.

**If a cell errors,** copy the red error text — most Colab issues here are
dependency/version tweaks that are quick to patch.
